```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

```python
path = r'.\data\states'
all_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
all_files
```

```python
census = pd.concat(map(pd.read_csv, all_files))
```

```python
census.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
census
```

```python
census.reset_index(drop=True, inplace=True)
display(census.describe())
display(census.info())
```

```python
census.dtypes
```

```python
census.head()
```

```python
census['Income'] = census['Income'].replace({'\$':''}, regex = True)
```

```python
census.head()
```

```python
census[['Male', 'Female']] = census['GenderPop'].str.split('_', 1, expand=True)
census.drop(['GenderPop'], axis=1, inplace=True)
census.head()
```

```python
census['Male'] = census['Male'].str.replace(r'M', '')
census['Female'] = census['Female'].str.replace(r'F', '')
```

```python
census.head()
```

```python
census['Hispanic'] = census['Hispanic'].str.replace(r'%', '')
census['White'] = census['White'].str.replace(r'%', '')
census['Black'] = census['Black'].str.replace(r'%', '')
census['Native'] = census['Native'].str.replace(r'%', '')
census['Asian'] = census['Asian'].str.replace(r'%', '')
census['Pacific'] = census['Pacific'].str.replace(r'%', '')
census.head()
```

```python
census_columns = census.columns[1:]
```

```python
census_columns
```

```python
for column in census_columns:
    census[column] = pd.to_numeric(census[column], errors='coerce')
```

```python
census[30:40]
```

```python
plt.scatter(census['Female'], census['Income'])
plt.show()
```

```python
census.isnull().sum()
```

```python
census['Female'].fillna(census['TotalPop']-census['Male'], inplace=True)
```

```python
census.isna().sum()
```

```python
census['Pacific'].fillna(
    100-census['Hispanic']-census['White']-census['Black']-census['Native']-census['Asian'],
    inplace=True
)
```

```python
census.isna().sum()
```

```python
census.duplicated().sum()
```

```python
census.drop_duplicates(inplace=True)
census.duplicated().sum()
```
